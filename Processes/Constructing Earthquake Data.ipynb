```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("earthquakes-data.tsv", sep='\t')
year_list = list(df[1:]['Year'])
```

```python
years = [i for i in range(1750,2023,1)]
year_count = []
for i in years:
    year_count.append(year_list.count(i))
```

```python
print(year_count)
```

```python
actual_df = pd.DataFrame(list(zip(years, year_count)), columns=['Year', 'Number of Earthquakes'])
actual_df.to_csv("Annual Number of Earthquakes (1750-2022)")
```

```python
actual_df.plot(x='Year', y=['Number of Earthquakes'])
```
