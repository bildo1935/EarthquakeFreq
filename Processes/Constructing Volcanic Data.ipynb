```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("volcano-data.tsv", sep='\t')
year_list = list(df[1:]['Year'])
```

```python
years = list(set(year_list))
year_count = []
counted_years = []
for i in year_list:
    if i not in counted_years:
        year_count.append(year_list.count(i)) 
        counted_years.append(i)
```

```python
for i in range(1750, 2022, 1):
    if i not in years:
        years.append(i)
        year_count.append(0)
```

```python
df1 = pd.DataFrame(list(zip(years, year_count)), columns=['Year', 'Number of Volcanic Eruptions'], dtype='int')
```

```python
df1.sort_values(by=['Year'], inplace=True)
df1.to_csv("Annual Number of Volcanic Eruptions (1750-2022)")
```

```python
df1.plot(x='Year', y=['Number of Volcanic Eruptions'])
```
